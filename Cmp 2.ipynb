```python
df_lea = pd.read_csv('la_and_region_codes_meta.csv')

```

```python
df.rename(columns = {'LEA_x':'LEA'}, inplace = True) 

df_8 = df.merge(df_lea, on='LEA', how='left')
```

```python
#question 8
df_8a = df_8[['LA Name','P8MEA','ATT8SCR']]
#df_8a.P8MEA.fillna(0)
df_8a = df_8a[df_8a.P8MEA != 'NE']
df_8a = df_8a[df_8a.P8MEA != 'SUPP']
df_8a = df_8a[df_8a.P8MEA != 'LOWCOV']
df_8a['P8MEA'] = df_8a['P8MEA'].astype(float)
g = df_8a.groupby('LA Name').agg({'P8MEA': ['mean']}) 
g = g.dropna()

res_h = g.apply(lambda x: x.sort_values(ascending=False).head(1))
res_l = g.apply(lambda x: x.sort_values(ascending=False).tail(1))
print("Local Authority with highest progress 8 scores")
print(res_h)
print("Local Authority with lowest progress 8 scores")
print(res_l)
g_minmax = df_8a.groupby('LA Name')['P8MEA'].agg(['max','min'])
g_minmax['diff'] = g_minmax['max']-g_minmax['min']
g_minmax1 = g_minmax['diff']
g_minmax1 = g_minmax1.to_frame()
g_minmax1 = g_minmax1.dropna()
range_h = g_minmax1.apply(lambda x: x.sort_values(ascending=False).head(1))
range_l = g_minmax1.apply(lambda x: x.sort_values(ascending=False).tail(1))
print("Local Authority with biggest range of values for progress 8 scores")
print(range_h)
print("Local Authority with lowest range of values for progress 8 scores")
print(range_l)
```

```python
df_8a['ATT8SCR'] = df_8a['ATT8SCR'].astype(float)
g= df_8a.groupby('LA Name').agg({'ATT8SCR': ['mean']}) 
g = g.dropna()

res_h = g.apply(lambda x: x.sort_values(ascending=False).head(1))
res_l = g.apply(lambda x: x.sort_values(ascending=False).tail(1))

print("Local Authority with highest attainment scores")
print(res_h)
print("Local Authority with lowest attainment scores")
print(res_l)
g_minmax = df_8a.groupby('LA Name')['ATT8SCR'].agg(['max','min'])
g_minmax['diff'] = g_minmax['max']-g_minmax['min']
g_minmax1 = g_minmax['diff']
g_minmax1 = g_minmax1.to_frame()
g_minmax1 = g_minmax1.dropna()
range_h = g_minmax1.apply(lambda x: x.sort_values(ascending=False).head(1))
range_l = g_minmax1.apply(lambda x: x.sort_values(ascending=False).tail(1))
print("Local Authority with biggest range of values for attainment scores")
print(range_h)
print("Local Authority with lowest range of values for attainment scores")
print(range_l)
```

```python
#question 9
df_9 = df_8[['ATT8SCR','REGION NAME']]
df_9 = df_9[df_9.ATT8SCR != 'NE']
df_9 = df_9[df_9.ATT8SCR != 'SUPP']
df_9['ATT8SCR'] = df_9['ATT8SCR'].astype(float)

print("Region with highest Attainment values")
print(df_9[df_9.ATT8SCR == df_9.ATT8SCR.max()])
print("Region with lowest Attainment values")
print(df_9[df_9.ATT8SCR == df_9.ATT8SCR.min()])
```
