```python
import pandas as pd
import os
```

```python
##Question1
#Reading excel file using pandas read_excel function
school_info = pd.read_excel('../england_school_information.xlsx')
#Filtering data for secondary schools
school_info = school_info [school_info.ISSECONDARY == 1]
#Filtering
school_info=school_info[['URN', 'LA','GENDER','RELCHAR']]
```

```python
##Question 2
cencus = pd.read_excel('../england_census.xlsx')
cencus['PNUMFSM'] = round(cencus['NUMFSM']/cencus['NOR']*100,1)
cencus=cencus[['URN', 'LA','PNORG','PNORB','PSENELSE','PSENELK','PNUMEAL', 'PNUMENGFL', 'PNUMUNCFL','NUMFSMEVER', 'PNUMFSMEVER']]
```

```python
##Question 3
abs_df = pd.read_excel('../england_abs.xlsx')
abs_df = abs_df.drop('ESTAB', 1)
```

```python
##Question 4
provisional = pd.read_excel('../england_ks4provisional.xlsx')
provisional = provisional[['URN', 'LEA', 'TPUP', 'P8PUP', 'ATT8SCR', 'P8MEA', 'P8CILOW', 'P8CIUPP', 'EBACCAPS', 'PTL2BASICS_94', 'PTL2BASICS_95', 'PTEBACC_E_PTQ_EE', 'PTEBACC_94', 'PTEBACC_95', 'PT5EM_94','P8_BANDING']]
```

```python
##Question 5
pupdes = pd.read_excel('../england_ks4-pupdest.xlsx')
pupdes = pupdes[['URN', 'LEA', 'OVERALL_DESTPER', 'NOT_SUSTAINEDPER', 'UNKNOWNPER', 'OVERALL_DESTPER_DIS','NOT_SUSTAINEDPER_DIS' ,'UNKNOWNPER_DIS']]
```

```python
##Question 6
school_info1 = school_info.merge(cencus, on='URN', how='left')
school_info1 = school_info1.merge(abs_df, on='URN', how='left')
school_info1 = school_info1.merge(provisional, on='URN', how='left')
school_info1 = school_info1.merge(pupdes, on='URN', how='left')
```

```python
#Question 7
ratings = pd.read_excel('../OFSTED Ratings.xlsx')
school_info1 = school_info1.merge(ratings, on='URN', how='left')
```
